```python
import matplotlib.pyplot as plt
import numpy as nump
 
element_ofx = nump.arange(0, 25, 4)
element_ofy = nump.arange(0, 26, 4)
 
# This numpy function creates 2-dimensional grid
[grid_ofX, grid_ofY] = nump.meshgrid(element_ofx, element_ofy)
 
# plots 2 graphs in one chart
fig, fillplot = plt.subplots(1, 1)
 
# Mathematical function for contour
grid_ofZ = nump.cos(grid_ofX / 1) - nump.sin(grid_ofY / 2)
 
# plots contour lines
fillplot.contourf(grid_ofX, grid_ofY, grid_ofZ)
 
fillplot.set_title('Contour Plot')
fillplot.set_xlabel('features of x-axis')
fillplot.set_ylabel('features of y-axis')
 
plt.show()
```

```python
#In civil engineering a contour plot could show the topoxlogy of a building site
#In mechanical engineering a contour plot could show the stress gradient across part surface
#Matplotlib's plt.contourf method
#Building contour plot with matplotlib entails using the ax.contour()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# where X and Y are 2D arrays of X and Y points; and Z is a 2D array of points that determine the height of the contour which is represented by color in a 2D plot. 

x=np.arange(-5,3,2)
y=np.arange(-5,3,2)
X,Y=np.meshgrid(x,y) 
Z=np.sign(X)*np.cos(Y)
fig,ax=plt.subplots(figsize=(6,6))
ax.contour(X,Y,Z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(-10,-5,2)
y=np.arange(-10,-5,2)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
fig,(ax1,ax2)=plt.subplots(1,2, figsize=(12,6))
mycmap1=plt.get_cmap("gist_earth")
ax1.set_aspect("equal")
ax1.set_title("Colormap:gist_earth")
cf1=ax1.contourf(X,Y,Z, cmap=mycmap1)
fig.colorbar(cf1,ax=ax1)
mycmap2=plt.get_cmap("gnuplot2")
ax2.set_aspect("equal")
ax2.set_title("Colormap:gnuplot2")
cf2=ax2.contourf(X,Y,Z, cmap=mycmap2)
fig.colorbar(cf2,ax=ax2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(-4,5,3)
y=np.arange(-5,5,3)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
fig,(ax_1,ax_2)=plt.subplots(1,2, figsize=(12,6))
mycmap_1=plt.get_cmap("Accent_r")
ax_1.set_aspect("equal")
ax_1.set_title("Colormap:Accent_r1")
cf_1=ax_1.contourf(X,Y,Z, cmap=mycmap_1)
fig.colorbar(cf_1,ax=ax_1)

mycmap_2=plt.get_cmap("twilight_shifted")
ax_2.set_aspect("equal")
ax_2.set_title("Colormap:magmar")
cf_2=ax_2.contourf(X,Y,Z, cmap=mycmap_2)
fig.colorbar(cf_2,ax=ax_2)
plt.show()
```
